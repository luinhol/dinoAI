```python
def relu(x):
	return max(0.0, x)
```

```python
def sigmoid(x):
	return 1/(1 + np.exp(-x))
```

```python
weights = np.random.rand(32)
print(weights)
```

```python
print(weights[28:32])
```

```python
entradas_neuronios_entrada = np.random.rand(7)
saida_neuronios_intermediarios = []
weights = np.random.rand(32)
```

```python
# para cada neuronio intermediario
for i in range(4):
    
    pesos = weights[i*7:i*7+7]
    
    saida_linear_intermediario = sum(x * y for x, y in zip(saida_neuronios_entrada, pesos))
    
    saida_ativacao_intermediario = relu(saida_linear_intermediario)
    
    saida_neuronios_intermediarios.append(saida_ativacao_intermediario)
    
pesos = weights[28:32]

saida_linear_saida = sum(x * y for x, y in zip(saida_ativacao_intermediario, pesos))
saida_ativacao_intermediario_saida = sigmoid(saida_linear_saida)
                         
if (saida_ativacao_intermediario_saida >= 0.55):
    return "K_UP"
else:
    return "K_DOWN"
```

```python
matriz_pesos_1_camada = weights[0:28].reshape(4,7)
matriz_pesos_2_camada = weights[28:32].reshape(1,4)

coluna_entradas = entradas_neuronios_entrada.reshape(7,1)

saida_linear_neuronios_intermediarios = np.dot(matriz_pesos_1_camada, coluna_entradas)
saida_ativacao_neuronios_intermediarios = np.maximum(0, saida_linear_neuronios_intermediarios)

print(saida_ativacao_neuronios_intermediarios)

saida_linear_neuronios_saida = np.dot(matriz_pesos_2_camada, saida_ativacao_neuronios_intermediarios)
print(saida_linear_neuronios_saida)

saida_ativacao_neuronios_saida = sigmoid(saida_linear_neuronios_saida)

print(saida_ativacao_neuronios_saida)
```
